```python
import os

# 指定文件夹路径和文件类型
folder_path = './'
file_extension = '.css'  # 根据你的需要修改文件类型，例如 .txt, .py

# 输出文件路径
output_file = 'all_css.txt'

# 创建或清空输出文件
with open(output_file, 'w', encoding='utf-8') as outfile:
    # 遍历文件夹及其子文件夹
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 过滤指定类型文件
            if file.endswith(file_extension):
                # 获取文件相对路径
                relative_path = os.path.relpath(os.path.join(root, file), folder_path)
                
                # 写入分隔符和相对路径
                outfile.write(f'===========\n{relative_path}\n===========\n')
                
                # 打开并读取文件内容
                with open(os.path.join(root, file), 'r', encoding='utf-8') as infile:
                    content = infile.read()
                    outfile.write(f'{content}\n')

print(f'所有文件已合并到 {output_file}')
```

```python
import os

def list_directory(path):
    for root, dirs, files in os.walk(path):
        # 打印当前目录路径
        print(f'当前目录: {root}')
        
        # 打印所有子目录
        print('子目录:')
        for dir_name in dirs:
            print(f'  - {dir_name}')
        
        # 打印所有文件
        print('文件:')
        for file_name in files:
            print(f'  - {file_name}')
        print('-' * 40)

# 使用绝对路径或相对路径指定文件夹
directory_path = './'
list_directory(directory_path)
```

```python
import os

def rename_images_to_png(folder_path):
    # 遍历文件夹中的所有文件
    for filename in os.listdir(folder_path):
        # 获取文件的完整路径
        file_path = os.path.join(folder_path, filename)
        
        # 检查是否为文件（排除文件夹）
        if os.path.isfile(file_path):
            # 分离文件名和扩展名
            name, ext = os.path.splitext(filename)
            
            # 忽略已经是 .png 的文件
            if ext.lower() != ".png":
                # 构造新的文件路径
                new_file_path = os.path.join(folder_path, name + ".png")
                
                # 重命名文件
                os.rename(file_path, new_file_path)
                print(f"Renamed: {filename} -> {name}.png")

# 使用示例
folder_path = "./images"  # 替换为实际文件夹路径
rename_images_to_png(folder_path)
```
